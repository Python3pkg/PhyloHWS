Here we define the parameters to be used. The parameters specific for a given application are included in arguments dictionary, while the other parameters are relative to the handling of the job 

```python
arguments={
"Nrandomization":"10",
"qparamHill":"1",
"output_prefix":"ciccio",
"XMLoutputType":"phyloxml"}
arguments["tree"]=open("tree.txt","r")
arguments["sample"]=open("sample.txt","r")
arguments["group"]=open("group.txt","r")

name="PhyloH"
mail="ciccio.bello@gmail.com"
session=str(int(random.random()*1000000))
```

Upload files  and Submit job and retrive job identifier

```python
JobID=UploadAndSubmitJST(arguments,name,session=session,mail=mail)
```
